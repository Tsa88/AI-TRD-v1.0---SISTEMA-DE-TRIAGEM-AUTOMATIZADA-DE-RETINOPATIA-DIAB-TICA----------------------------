def gerar_dados_sinteticos_calon(num_pacientes=500):
    """
    Gera um dataset (banco de dados) fictício em conformidade com a LGPD
    para simular a validação do modelo preditivo no SUS.
    """
    print("[1/4] Gerando coorte sintética de pacientes (Mock Data)...")
    np.random.seed(42) # Garante que o teste seja reprodutível
    
    # Variáveis Clínicas e Sociais (0 = Não/Baixo, 1 = Sim/Alto)
    idades = np.random.randint(18, 65, num_pacientes)
    hpv_positivo = np.random.choice([0, 1], num_pacientes, p=[0.7, 0.3])
    historico_papanicolau = np.random.choice([0, 1], num_pacientes, p=[0.4, 0.6])
    
    # Determinantes Sociais da Saúde (Foco na população Calon)
    transitoriedade_geografica = np.random.choice([0, 1], num_pacientes, p=[0.5, 0.5])
    acesso_regular_ubs = np.where(transitoriedade_geografica == 1,
                                  np.random.choice([0, 1], num_pacientes, p=[0.8, 0.2]),
                                  np.random.choice([0, 1], num_pacientes, p=[0.3, 0.7]))
    
    # Lógica base do Risco Oncológico (Alvo preditivo)
    # Se tem HPV, não faz Papanicolau e tem alta transitoriedade, o risco dispara.
    risco_base = (hpv_positivo * 3) + ((1 - historico_papanicolau) * 2) + (transitoriedade_geografica * 1.5) - (acesso_regular_ubs * 1)
    risco_oncologico = np.where(risco_base >= 3.5, 1, 0) # 1 = Alto Risco, 0 = Baixo/Médio Risco
    
    df_pacientes_calon_anonimizado = pd.DataFrame({
        'idade_paciente': idades,
        'status_hpv': hpv_positivo,
        'historico_preventivo': historico_papanicolau,
        'fator_itinerancia': transitoriedade_geografica,
        'vinculo_ubs': acesso_regular_ubs,
        'diagnostico_risco_alto': risco_oncologico
    })
    
    return df_pacientes_calon_anonimizado

def treinar_modelo_sus(df_dados):
    """
    Treina o classificador Random Forest com os dados clínicos.
    """
    print("[2/4] Preparando variáveis (Features) e Alvo (Target)...")
    
    # X = Tudo o que o modelo usa para aprender (A "Gasolina")
    X_features = df_dados.drop('diagnostico_risco_alto', axis=1)
    # y = O que o modelo precisa prever (O "Destino")
    y_target = df_dados['diagnostico_risco_alto']
    
    # Divide os dados: 80% para treinar o algoritmo, 20% para testar se ele aprendeu
    X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size=0.2, random_state=42)
    
    print("[3/4] Inicializando o Motor Preditivo (Random Forest)...")
    classificador_oncologico_rf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
    
    # Treinando o modelo
    classificador_oncologico_rf.fit(X_train, y_train)
    
    return classificador_oncologico_rf, X_test, y_test

def avaliar_desempenho_clinico(modelo, X_test, y_test):
    """
    Estratifica o risco e avalia a precisão da Inteligência Artificial.
    """
    print("[4/4] Gerando Estratificação de Risco e Relatório Final...\n")
    
    # A IA tenta adivinhar o risco das pacientes de teste
    risco_predito_sus = modelo.predict(X_test)
    
    acuracia = accuracy_score(y_test, risco_predito_sus)
    relatorio = classification_report(y_test, risco_predito_sus, target_names=['Baixo/Médio Risco (0)', 'Alto Risco (1)'])
    
    print("=====================================================")
    print(f" RELATÓRIO DE VALIDAÇÃO DA IA - CERVIX-PRED")
    print("=====================================================")
    print(f"Acurácia Geral do Algoritmo: {acuracia * 100:.2f}%\n")
    print("Detalhamento Clínico:")
    print(relatorio)
    print("=====================================================")
    print("Status: PRONTO PARA IMPLEMENTAÇÃO NO AMBIENTE SUS.")
